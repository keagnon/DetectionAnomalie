```python
def batch_merge(df_courbe, df_carburant, batch_size=100000):
    merged_batches = []  # Liste pour stocker les résultats intermédiaires

    # Calcul du nombre total de lignes dans df_courbe
    total_rows = df_courbe.shape[0]

    # Parcours de df_courbe par batchs
    for start_row in range(0, total_rows, batch_size):
        end_row = min(start_row + batch_size, total_rows)
        df_courbe_batch = df_courbe.iloc[start_row:end_row]

        # Fusion du batch actuel avec df_carburant
        merged_batch = pd.merge(df_courbe_batch, df_carburant, on='région', how='inner')

        # Ajout du résultat intermédiaire à la liste
        merged_batches.append(merged_batch)

        # Optionnel : Afficher la progression
        print(f'Batch {start_row//batch_size + 1} sur {((total_rows - 1)//batch_size) + 1} fusionné avec succès.')

    # Concaténation de tous les batchs fusionnés pour obtenir le DataFrame final
    merged_df = pd.concat(merged_batches, ignore_index=True)

    return merged_df

# Utilisation de la fonction
merged_df = batch_merge(df_courbe, df_carburant, batch_size=100000)

# Vérification du résultat
print(f'Taille du DataFrame fusionné : {merged_df.shape}')
```

```python
merged_df.to_csv('merged_courbe_carburants.csv',index=False, sep=';')
```
